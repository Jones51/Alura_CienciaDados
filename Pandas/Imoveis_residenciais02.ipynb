```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'.
selecao_apartamento = dados.Tipo == "Apartamento"
n1 = dados[selecao_apartamento]
n1.head(5)
```

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao_apartamentos = (dados.Tipo == "Casa") | (dados.Tipo == "Casa de Condomínio") | (dados.Tipo == "Casa de Vila") 
n2 = dados[selecao_apartamentos]
n2.head(5)
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
selecao_apartamentos = (dados.Area >= 60) & (dados.Area <= 100)
n3 = dados[selecao_apartamentos]
n3.head(5)
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao_apartamentos = (dados.Quartos >= 4) & (dados.Valor <= 2000.00)
n4 = dados[selecao_apartamentos]
n4.head(5)
```
